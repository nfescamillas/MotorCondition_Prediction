```python
import os 
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import Xception 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
# print("Num GPUs Available:", len(tf.config.list_physical_devices('GPU')))
```

```python
def load_images(base_dir, target_size):
    images=[]
    labels=[]
    class_names = sorted(os.listdir(base_dir))
    class_names = [ name for name in class_names if name != 'Readme.txt']
    class_mapping ={name :idx for idx,name in enumerate(class_names)}

    for name in class_names:
        folder_path = os.path.join(base_dir, name)
        if os.path.isdir(folder_path):
            for file in os.listdir(folder_path):
                if file.endswith('bmp'):
                    img_path = os.path.join(folder_path,file)
                    img =Image.open(img_path)
                    img = img.resize((target_size))
                    img = np.asarray(img)
                    
                    images.append(img)
                    labels.append(class_mapping[name])
                
    return np.array(images), np.array(labels), class_mapping

```

```python
base_dir = './data' 
target_size = (150,150)

images, labels, class_mapping =load_images(base_dir, target_size)
images = images.astype('float32')/ 255
```

```python
base_model =Xception(weights ='imagenet',include_top=False, input_shape=(150,150,3))
base_model.trainable= False
```

```python
def model_arch(model,learning_rate,dense_layers=100, drop_rate=0):
    inputs = keras.Input(shape=(150,150,3))
    base = model(inputs)
    x = keras.layers.GlobalAveragePooling2D()(base)
    x =keras.layers.Dense(dense_layers,activation='relu')(x)
    x = keras.layers.Dropout(drop_rate)(x)
    outputs =keras.layers.Dense(len(class_mapping), activation='softmax')(x)
    model =keras.Model(inputs,outputs)
    optimizer =keras.optimizers.Adam(learning_rate=learning_rate)
    loss =keras.losses.CategoricalCrossentropy(from_logits=False)
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    return model 
```

```python
train_image,test_image,train_labels, test_labels = train_test_split(images, labels, test_size=0.2, shuffle=True, random_state=42)
train_labels = to_categorical(train_labels, num_classes=11)
test_labels = to_categorical(test_labels, num_classes=11)
```

```python
scores ={}

for lr in [0.0001,0.001,0.01,0.1]:
    model = model_arch(base_model,lr)
    history =model.fit(train_image, train_labels, epochs=10, batch_size=8, validation_split =0.2)
    scores[lr] =history.history
```

```python
for lr,hist in scores.items():
    plt.plot(hist['val_accuracy'],label=lr)
plt.legend()
plt.show()
```

```python
scores ={}
lr=0.01
for size in [50,100,150,200]:
    model = model_arch(base_model,lr,dense_layers=size)
    history =model.fit(train_image, train_labels, epochs=10, batch_size=8, validation_split =0.2)
    scores[size] =history.history
```

```python
for size,hist in scores.items():
    plt.plot(hist['val_accuracy'],label=size)
plt.legend()
plt.show()
```

```python
scores ={}
lr=0.01
size=100

for rate in [0,0.2,0.4,0.6]:
    model = model_arch(base_model,lr,dense_layers=size,drop_rate =rate)
    history =model.fit(train_image, train_labels, epochs=10, batch_size=8, validation_split =0.2)
    scores[rate] =history.history
```

```python
for rate,hist in scores.items():
    plt.plot(hist['val_accuracy'],label=rate)
plt.legend()
plt.show()
```

```python
checkpoint =keras.callbacks.ModelCheckpoint('xception_v3_{epoch:02d}_{val_accuracy:.3f}.h5',
                                            save_best_only= True,
                                            monitor ='val_accuracy',
                                            mode = 'max')
```

```python
learning_rate= 0.01
dense_layers=100
rate=0

model = model_arch(base_model,learning_rate,dense_layers,rate)
history =model.fit(train_image, train_labels, epochs=10, batch_size=8, validation_split =0.2,callbacks =[checkpoint])
```

```python
plt.plot(history.history['accuracy'],label="train")
plt.plot(history.history['val_accuracy'],label="test")
plt.legend()
plt.show()
```
