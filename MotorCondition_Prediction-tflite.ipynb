```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
tf.__version__
```

```python
model =keras.models.load_model('xception_v3_07_0.966.h5')
```

```python
path = '/home/nikki/ML_MotorIRProject/data/A_B50/298.bmp'
```

```python
img= load_img(path, target_size=(150,150))
```

```python
plt.imshow(img)
plt.show()
```

```python
img = np.asarray(img)
img = np.array([img])
img = img.astype('float32') # needed float32 due to interpreter invoking in tflite
```

```python
classes =[
    'A10',
 'A30',
 'A50',
 'A_B50',
 'A_C10',
 'A_C30',
 'A_C_B10',
 'A_C_B30',
 'Fan',
 'Noload',
 'Rotor_0']
```

```python
preds =model.predict(img)
```

```python
dict(zip(classes,preds[0]))
```

### CONVERT KERAS TO TF LITE

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open('motorir_model.tflite','wb') as f_out:
    f_out.write(tflite_model)
```

```python
import tensorflow.lite as tflite
```

```python
interpreter =tflite.Interpreter(model_path='motorir_model.tflite')
interpreter.allocate_tensors()
```

```python
interpreter.get_input_details()
```

```python
input_index= interpreter.get_input_details()[0]['index']
```

```python
interpreter.get_output_details()
```

```python
output_index =interpreter.get_output_details()[0]['index']
```

```python
interpreter.set_tensor(input_index, img)
interpreter.invoke()
```

```python
preds =interpreter.get_tensor(output_index)
```

```python
preds
```
